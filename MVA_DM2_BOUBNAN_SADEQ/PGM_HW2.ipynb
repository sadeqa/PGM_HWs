# Gaussian mixtures models

```python
from PGM_HW2_tools import my_GMM
from PGM_HW2_tools import my_iso_GMM
from PGM_HW2_tools import plot_results
```

### Fitting gaussian mixture models on train data

Fitting an **Isotropic** Gaussian mixture model with 3 clusters on `train data`

```python
GMM_iso=my_iso_GMM(n_clusters=4)
GMM_iso.fit(X_data)
```

Fitting a **General** Gaussian mixture model with 3 clusters on `train data`

```python
GMM=my_GMM(n_clusters=4)
GMM.fit(X_data)
```

We plot the results :

```python
plot_results(X=X_data, iso_model=GMM_iso,iso_labels=GMM_iso.labels_, general_model=GMM,general_labels=GMM.labels_, data_label="Train fit", confidence=0.90)
```

### Predicting the test data labels

We'll use the **fitted** gaussian mixture models to `predict` the test data labels :

```python
Test_iso_labels = GMM_iso.predict(X_test)
Test_general_labels = GMM.predict(X_test)
```

```python
plot_results(X=X_test, iso_model=GMM_iso, iso_labels=Test_iso_labels, general_model=GMM, general_labels=Test_general_labels, data_label="Test predict", confidence=0.90)
```
