```python
import numpy as np 
import matplotlib.pyplot as plt 

''' Dataset A '''
x_train_A = np.loadtxt('classification_data_HWK1/classificationA.train')[:,:-1]
y_train_A = np.loadtxt('classification_data_HWK1/classificationA.train')[:,-1].reshape(-1,1)

x_test_A = np.loadtxt('classification_data_HWK1/classificationA.test')[:,:-1]
y_test_A = np.loadtxt('classification_data_HWK1/classificationA.test')[:,-1].reshape(-1,1)


''' Dataset B '''
x_train_B = np.loadtxt('classification_data_HWK1/classificationB.train')[:,:-1]
y_train_B = np.loadtxt('classification_data_HWK1/classificationB.train')[:,-1].reshape(-1,1)

x_test_B = np.loadtxt('classification_data_HWK1/classificationB.test')[:,:-1]
y_test_B = np.loadtxt('classification_data_HWK1/classificationB.test')[:,-1].reshape(-1,1)


''' Dataset C '''
x_train_C = np.loadtxt('classification_data_HWK1/classificationC.train')[:,:-1]
y_train_C = np.loadtxt('classification_data_HWK1/classificationC.train')[:,-1].reshape(-1,1)

x_test_C = np.loadtxt('classification_data_HWK1/classificationC.test')[:,:-1]
y_test_C = np.loadtxt('classification_data_HWK1/classificationC.test')[:,-1].reshape(-1,1)
```

```python
f, ((ax11,ax12),(ax21,ax22),(ax31,ax32)) = plt.subplots(3,2, figsize=(12,12))

ax11.scatter(x_train_A[:,0],x_train_A[:,1],c=y_train_A.flatten())
ax11.set_title("Train set A")
ax12.scatter(x_test_A[:,0],x_test_A[:,1],c=y_test_A.flatten())
ax12.set_title("Test set A")

ax21.scatter(x_train_B[:,0],x_train_B[:,1],c=y_train_B.flatten())
ax21.set_title("Train set B")
ax22.scatter(x_test_B[:,0],x_test_B[:,1],c=y_test_B.flatten())
ax22.set_title("Test set B")

ax31.scatter(x_train_C[:,0],x_train_C[:,1],c=y_train_C.flatten())
ax31.set_title("Train set C")
ax32.scatter(x_test_C[:,0],x_test_C[:,1],c=y_test_C.flatten())
ax32.set_title("Test set C")

plt.tight_layout()
plt.show()
```

# Linear Discriminant Analysis

The assumption we consider in our model :

- $P(X/Y=1) \sim \mathcal{N}(\mu_1,\Sigma_1)$ and $P(X/Y=0) \sim \mathcal{N}(\mu_0,\Sigma_0)$
- $P(Y) \sim \mathcal{B(\pi)}$
- Fisher's assumption : $\Sigma_0=\Sigma_1=\Sigma$

```python
from PGM_HW_tools import LinearDiscriminantAnalysis
from PGM_HW_tools import plot_contour_LDA
```

```python
LDA = LinearDiscriminantAnalysis()
LDA.fit(x_train_A,y_train_A)
plot_contour_LDA(LDA,'A',x_train_A,y_train_A,x_test_A,y_test_A)

LDA = LinearDiscriminantAnalysis()
LDA.fit(x_train_B,y_train_B)
plot_contour_LDA(LDA,'B',x_train_B,y_train_B,x_test_B,y_test_B)

LDA = LinearDiscriminantAnalysis()
LDA.fit(x_train_C,y_train_C)
plot_contour_LDA(LDA,'C',x_train_C,y_train_C,x_test_C,y_test_C)
```
